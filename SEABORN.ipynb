```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(tips['total_bill'],kde=False,color='red',bins=50)
plt.title("distribution plot")
plt.xlabel=("total bill amount")
plt.ylabel=("frequency")
plt.show()
```

```python
sns.jointplot(x=tips["total_bill"],y=tips["tip"],kind='hex',color="purple")
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
plt.show()
```
