```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification

# Create imbalanced dataset (90% class 0, 10% class 1)
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0, 
                          n_informative=2, random_state=42, 
                          n_clusters_per_class=1, weights=[0.9, 0.1])

# Plot the dataset
plt.figure(figsize=(10, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], alpha=0.7, label='Class 0 (Majority)', color='blue')
plt.scatter(X[y==1, 0], X[y==1, 1], alpha=0.7, label='Class 1 (Minority)', color='red')
plt.title('Visualization of Class Imbalance (90% Class 0, 10% Class 1)', fontsize=14)
plt.xlabel('Feature 1', fontsize=12)
plt.ylabel('Feature 2', fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Create pie chart showing class distribution
plt.figure(figsize=(8, 8))
labels = ['Class 0 (Majority)', 'Class 1 (Minority)']
sizes = [np.sum(y==0), np.sum(y==1)]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.title('Class Distribution', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification

# Create imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                          n_informative=2, random_state=42,
                          n_clusters_per_class=1, weights=[0.9, 0.1])

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
df['Class'] = y

# Check class distribution
class_counts = df['Class'].value_counts()
print("Class Distribution:")
print(class_counts)
print(f"Imbalance Ratio: 1:{class_counts[0]/class_counts[1]:.1f}")

# Visualize class distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution in Dataset', fontsize=14)
plt.xlabel('Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
for i, count in enumerate(class_counts):
    plt.text(i, count + 5, f"{count} ({count/len(df)*100:.1f}%)", 
             ha='center', fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (confusion_matrix, precision_recall_curve, 
                            roc_curve, roc_auc_score, precision_recall_fscore_support,
                            average_precision_score)
import seaborn as sns

# Create imbalanced dataset
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
                          n_informative=2, random_state=42,
                          n_clusters_per_class=1, weights=[0.95, 0.05])

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a standard logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Get predictions and probabilities
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# Calculate various metrics
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
accuracy = (y_pred == y_test).mean()
auc_roc = roc_auc_score(y_test, y_proba)
auc_pr = average_precision_score(y_test, y_proba)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix', fontsize=14)
plt.xlabel('Predicted', fontsize=12)
plt.ylabel('Actual', fontsize=12)
plt.show()

# Plot ROC Curve
plt.figure(figsize=(10, 6))
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr, lw=2, label=f'ROC curve (AUC = {auc_roc:.3f})')
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14)
plt.legend(loc="lower right", fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Plot Precision-Recall Curve
plt.figure(figsize=(10, 6))
precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_proba)
plt.plot(recall_curve, precision_curve, lw=2, label=f'PR curve (AP = {auc_pr:.3f})')
plt.axhline(y=sum(y_test)/len(y_test), color='r', linestyle='--', 
           label=f'Baseline (class ratio = {sum(y_test)/len(y_test):.3f})')
plt.xlabel('Recall', fontsize=12)
plt.ylabel('Precision', fontsize=12)
plt.title('Precision-Recall Curve', fontsize=14)
plt.legend(loc="best", fontsize=12)
plt.grid(alpha=0.3)
plt.show()

# Display all metrics
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 Score: {f1:.3f}")
print(f"AUC-ROC: {auc_roc:.3f}")
print(f"AUC-PR: {auc_pr:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler

# Create imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                          n_informative=2, random_state=42,
                          n_clusters_per_class=1, weights=[0.9, 0.1])

# Apply random undersampling
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Original data
axes[0].scatter(X[y==0, 0], X[y==0, 1], alpha=0.7, label='Class 0 (Majority)', color='blue')
axes[0].scatter(X[y==1, 0], X[y==1, 1], alpha=0.7, label='Class 1 (Minority)', color='red')
axes[0].set_title('Original Imbalanced Dataset', fontsize=14)
axes[0].set_xlabel('Feature 1', fontsize=12)
axes[0].set_ylabel('Feature 2', fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# Undersampled data
axes[1].scatter(X_resampled[y_resampled==0, 0], X_resampled[y_resampled==0, 1], 
               alpha=0.7, label='Class 0 (Undersampled)', color='blue')
axes[1].scatter(X_resampled[y_resampled==1, 0], X_resampled[y_resampled==1, 1], 
               alpha=0.7, label='Class 1', color='red')
axes[1].set_title('After Random Undersampling', fontsize=14)
axes[1].set_xlabel('Feature 1', fontsize=12)
axes[1].set_ylabel('Feature 2', fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Print class distribution
print(f"Original class distribution: {np.bincount(y)}")
print(f"Resampled class distribution: {np.bincount(y_resampled)}")
```
